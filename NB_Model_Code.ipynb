# **Checking, If Any Column Contains Null Values**


```python
for col in df:
  print(col.upper(), "Column contains: \n")
  print(df[col].value_counts(),'\n\n\n')
 
```

```python
#Replacing '?' with 'NaN' values
df.replace('?', np.NaN, inplace = True)
df.head()
```

```python
#Applying Forward Fill to Replace 'NaN'
df.fillna(method = 'ffill', inplace = True)
df.head()
```

Here we found that
Workclass column has 1376 values from unknown category.
Occupation column has 1381 values from unknown category.
Native Country column has 449 values from unknown country.

```python
#Dropping Unuseful data
df.drop(['education.num', 'fnlwgt','capital.gain', 'capital.loss'], axis = 1, inplace=True)
df.head()
```

We Assumed that educatio.num, fnlwgt, capital.gain, capital.loss columns are not affecting our data more so we drop these.

# **Lab**el **Encoding**
Here we will apply label encoding for changing categorical values into numerical because machine understands only numbers.

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['workclass'] = le.fit_transform(df['workclass'])
df['marital.status'] = le.fit_transform(df['marital.status'])
df['occupation'] = le.fit_transform(df['occupation'])
df['relationship'] = le.fit_transform(df['relationship'])
df['sex'] = le.fit_transform(df['sex'])
df['race'] = le.fit_transform(df['race'])
df['native.country'] = le.fit_transform(df['native.country'])
df['income'] = le.fit_transform(df['income'])
df.head()
```

```python
df = df.apply(LabelEncoder().fit_transform)
df.head()
```

```python
x = df.drop(['income'], axis = 1)   #Dropping column ['income'] 
y = df['income']    #Assigning column ['income'] to y or using as a class
```

# Data Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2) #70% data as training_data and 30% as testing_data
```

```python
x_train.info()
```

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
gnb = GaussianNB() #Creating object for Gaussian Naive Bayes
gnb.fit(x_train, y_train) #fitting data into Gaussian Naive Bayes

bnb = BernoulliNB() #Creating object for Bernoulli Naive Bayes
bnb.fit(x_train, y_train) #fitting data into Bernoulli Naive Bayes

mnb = MultinomialNB() #Creating object for Multinomial Naive Bayes
mnb.fit(x_train, y_train) #fitting data into Multinomial Naive Bayes
```

```python
#Performance Computing modules
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
#Gausssian Naive Bayes
y_pred = gnb.predict(x_test)
print("Classification Report: \n",classification_report(y_test, y_pred))
print("Confusion Matrix is:\n ",confusion_matrix(y_test, y_pred))
print("Accuracy is: ",accuracy_score(y_test, y_pred))
```

```python
#Bernoulli Naive Bayes
y_pred = bnb.predict(x_test)
print("Classification Report:\n",classification_report(y_test, y_pred))
print("Confusion Matrix is:\n ",confusion_matrix(y_test, y_pred))
print("Accuracy is: ",accuracy_score(y_test, y_pred))
```

```python
#Multinomial Naive Bayes
y_pred = mnb.predict(x_test)
print("Classification Report:\n ",classification_report(y_test, y_pred))
print("Confusion Matrix is:\n ",confusion_matrix(y_test, y_pred))
print("Accuracy is: ",accuracy_score(y_test, y_pred))
```
